```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt

# Load data
results = pd.read_csv("/home/nnieto/Nico/Harmonization/QC/output/sex_classification/results_single_site_SALD_eNKI_CamCAN_all_sampling_Q.csv")
metric_to_plot = "AUC"

results_test = results[results["Model"] == "None Test"]

plt.figure(figsize=(15, 10))
sbn.boxplot(data=results_test, x="Site", y=metric_to_plot, hue="QC_Sampling")
plt.grid()
plt.title("AUC for single site classification")
```
